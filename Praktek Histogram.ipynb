```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt


img = cv2.imread('Kucing.jpg')

color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256]) 
    plt.plot(histr,color = col) 
    plt.xlim([0,256])
plt.show( )
```

```python
img = mpimg.imread('Kucing.jpg')
img_gray =cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gray_2 =cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB)
plt.hist(img_gray_2.ravel(),256,[0,256]);
```

```python
alpha = 2
adjusted_alpha = cv2.convertScaleAbs(img_gray_2, alpha=alpha)
plt.imshow(adjusted_alpha)
```

```python
adjusted_beta = cv2.convertScaleAbs(img_gray_2, beta=20)
plt.imshow(adjusted_beta)
```

```python
adjusted = cv2.convertScaleAbs(img_gray_2, alpha=1.2, beta=40)
plt.imshow(adjusted)
```
